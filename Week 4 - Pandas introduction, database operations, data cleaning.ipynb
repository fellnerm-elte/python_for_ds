```python
#reading from csv

df = pd.read_csv("data/IMDB-Movie-Data.csv", index_col="Title")
df.head()
```

```python
#getting information about the data

df.info()
df.shape
```

```python
#creating duplications

temp_df = df.append(df)

temp_df.shape
```

```python
#removing duplications

temp_df = temp_df.drop_duplicates()

temp_df.shape

temp_df.drop_duplicates(inplace=True)
```

```python
#keep

temp_df = df.append(df)  # make a new copy

temp_df.drop_duplicates(inplace=True, keep=False) #first, last

temp_df.shape
```

```python
#column cleaning

df.columns
```

```python
df.rename(columns={
        'Runtime (Minutes)': 'Runtime', 
        'Revenue (Millions)': 'Revenue'
    }, inplace=True)

df.columns
```

```python
#lowercasing

df.columns = ['rank', 'genre', 'description', 'director', 'actors', 'year',
       'runtime', 'rating', 'votes', 'revenue',
       'metascore']

df.columns
```

```python
#quicker method of lowercasing

df.columns = [col.lower() for col in df]
df.columns
```

```python
#DATA CLEANING
#Dropping columns

df = pd.read_csv('data/BL-Flickr-Images-Book.csv')
df.head()
```

```python
to_drop = ['Edition Statement',
           'Corporate Author',
           'Corporate Contributors',
           'Former owner',
           'Engraver',
           'Contributors',
           'Issuance type',
           'Shelfmarks']

df.drop(to_drop, inplace=True, axis=1)
#df.drop(columns=to_drop, inplace=True)
```

```python
df.head()
```

```python
#index changing
df['Identifier'].is_unique
df = df.set_index('Identifier')
df.head()
```

```python
#field tidying

extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```
